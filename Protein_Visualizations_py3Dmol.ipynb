```python
#lysozyme cartoon style
cif_data = open("4NHP.cif").read()

view = py3Dmol.view(width=400, height=400)

view.addModel(cif_data, "cif")

view.setStyle({'cartoon' : {'color' : 'spectrum'}})
view.zoomTo()
view.show()
```

```python
#hemoglobin stick and ball style
cif_data = open("1FDH.cif").read()

view = py3Dmol.view(width=400, height=400)

view.addModel(cif_data, "cif")

view.setStyle(
    {'sphere': {'radius': 0.3},
    'stick': {'radius': 0.2}})
view.zoomTo()
view.show()
```

```python
#hemoglobin cartoon style. no beta sheetsv
cif_data = open("1FDH.cif").read()

view = py3Dmol.view(width=500, height=500)

view.addModel(cif_data, "cif")

view.setStyle({'cartoon': {'color': 'spectrum', 'arrows' : True}})
view.zoomTo()
view.show()
```
